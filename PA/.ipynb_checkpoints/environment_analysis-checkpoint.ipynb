```python
resid_col = ['1B_resid', '2B_resid', '3B_resid', 'HR_resid', 'FO_resid', 'BB_resid', 'K_resid', 'other_resid']
pf_resid_col = ['1B_pf_resid', '2B_pf_resid', '3B_pf_resid', 'HR_pf_resid', 'FO_pf_resid', 'BB_pf_resid', 'K_pf_resid', 'other_pf_resid']
event_list = ['1B', '2B', '3B', 'HR', 'FO', 'BB', 'K', 'other']
```

```python
for e in event_list:
    all_matchUp_df[e+'_pf_resid'] = all_matchUp_df[e] - all_matchUp_df[e+'_proba_adj']
```

# Temp

```python
resid_decile_temp = all_matchUp_df.groupby(pd.qcut(all_matchUp_df['temp'], 10)).agg({
    '1B_resid':'mean',
    '2B_resid':'mean',
    '3B_resid':'mean',
    'HR_resid':'mean',
    'FO_resid':'mean',
    'K_resid':'mean',
    'BB_resid':'mean',
    'other_pf_resid':'mean',
    'other_resid':'mean',
    '1B_pf_resid':'mean',
    '2B_pf_resid':'mean',
    '3B_pf_resid':'mean',
    'HR_pf_resid':'mean',
    'FO_pf_resid':'mean',
    'K_pf_resid':'mean',
    'BB_pf_resid':'mean',
    'other_pf_resid':'mean'
}).reset_index().copy()
resid_decile_temp['deciles'] = resid_decile_temp.groupby(['temp']).ngroup()

resid_decile_temp
```

```python
fig, axes = plt.subplots(4, 2, figsize=[12, 12])
plt.tight_layout(pad=5)

for col, ax in zip(resid_col, axes.ravel()):
    sns.lineplot(data=resid_decile_temp, x='deciles', y=col, ax=ax)
    ax.set_xticks(range(10))
fig.suptitle('Residuals vs Temp Deciles', fontsize=20)
plt.show()
```

```python
fig, axes = plt.subplots(4, 2, figsize=[12, 12])
plt.tight_layout(pad=5)

for col, ax in zip(pf_resid_col, axes.ravel()):
    sns.lineplot(data=resid_decile_temp, x='deciles', y=col, ax=ax)
    ax.set_xticks(range(10))
fig.suptitle('PF Residuals vs Temp Deciles', fontsize=20)
plt.show()
```

## HR

```python
regr_hr = RandomForestRegressor(random_state=0)
```

```python
X_col_hr = ['HR_pf']
y_col_hr = ['HR_resid']
```

```python
X_train_hr = train_df[X_col_hr].values
y_train_hr = train_df[y_col_hr].values.flatten()
```

```python
X_test_hr = test_df[X_col_hr].values
y_test_hr = test_df[y_col_hr].values.flatten()
```

```python
rf_def_hr = regr_hr.fit(X_train_hr, y_train_hr)
```

```python
y_pred_hr = rf_def_hr.predict(X_test_hr)
```

```python
y_test_hr
```

```python
y_pred_hr
```

```python
print(f"Explained Variance: {explained_variance_score(y_test_hr, y_pred_hr)}")
print(f"MAE: {mean_absolute_error(y_test_hr, y_pred_hr)}")
print(f"MSE: {mean_squared_error(y_test_hr, y_pred_hr)}")
```
